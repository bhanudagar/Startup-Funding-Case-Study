## Question 2
Even after trying for so many times, your friend’s startup could not find the investment. So you decided to take this matter in your hand and try to find the list of investors who probably can invest in your friend’s startup. Your list will increase the chance of your friend startup getting some initial investment by contacting these investors. Find the top 5 investors who have invested maximum number of times (consider repeat investments in one company also). In a startup, multiple investors might have invested. So consider each investor for that startup. Ignore undisclosed investors.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("startup_funding.csv")
#dealing with NA entries in Investors Name
df.dropna(subset=["InvestorsName"], how="all",inplace=True)
#creaiting np array of all Investor's name
investorName = np.array(df["InvestorsName"])

#creating dictionary of investor:timesInvested
d = {}
for i in investorName:
    name = i.split(',')
    for j in name:
        j = j.strip()
        if j == '' or j == ' ' or "Undisclosed" in j or "undisclosed" in j:
            continue
        d[j] = d.get(j, 0) + 1
#sorting the dictionary on the bases of number of times invested in decreasing order
ans = sorted(d.items(), key = lambda x: x[1], reverse = True)[:5]

#Creating investors array and number of times array to visualize the data
investors = []
value = []

for i in ans:
    print(i[0], i[1])
    investors.append(i[0])
    value.append(i[1])
    
#VISUALIZATION
#bar graph
x = np.arange(5)
plt.bar(x, value, color = "Orange", width = 0.5)
plt.grid()
plt.xticks(x, investors, rotation = 40)
plt.xlabel("Name of Investor")
plt.ylabel("Number of times Invested")
plt.title("Top 5 investors in India")
plt.show()

#Pie chart
# plt.pie(value, labels = investors, colors = ["darkcyan", "pink", "mediumseagreen", "deepskyblue", "firebrick"], explode = [0.1, 0,0,0,0], autopct = "%.2f %%")
# plt.axis("equal")
# plt.title("Percentage of numer of times each Investor invested")
# plt.show()
```
