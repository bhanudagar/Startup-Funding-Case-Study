## Scenario 7 - Funding Rounds

Given File 'startup_funding.csv'
#### Problem Statement :
1. Find the top 5 startups who received the most number of funding rounds. That means, startups which got fundings maximum number of times.
2. Print the startup name in descending order with respect to the number of funding round as integer value.
#### Note:
1. Ola, Flipkart, Oyo, Paytm are important startups, so correct their names. There are many errors in startup names, ignore correcting all, just handle important ones.

### Output Format :
#### city1 percent1
#### city2 percent2
#### city3 percent3
#### . . . 
#### . . .
#### . . .






```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('startup_funding.csv')
df.head()
```

```python
df.StartupName.fillna('', inplace = True)

startup_funding = {}
for i in range(len(df.StartupName)):
    currName = df.StartupName[i]
    startup_funding[currName] = startup_funding.get(currName, 0) + 1
    
startup_funding
```

```python
ans = sorted(startup_funding.items(), key = lambda x : x[1], reverse = True)
ans
```

### As we already know from the previous question, we need to clean our data. Basically, we need to correct the spellings of top startups. Let's do this

```python
df.StartupName.replace('Ola Cabs', 'Ola', inplace = True)
df.StartupName.replace('Olacabs', 'Ola', inplace = True)
df.StartupName.replace('Oyo Rooms', 'Oyo', inplace = True)
df.StartupName.replace('Paytm Marketplace', 'Paytm', inplace = True)
df.StartupName.replace('Oyorooms', 'Oyo', inplace = True)
df.StartupName.replace('OyoRooms', 'Oyo', inplace = True)
df.StartupName.replace('OYO Rooms', 'Oyo', inplace = True)
df.StartupName.replace('Flipkart.com', 'Flipkart', inplace = True)
```

```python
startup_funding = {}
for i in range(len(df.StartupName)):
    currName = df.StartupName[i]
    startup_funding[currName] = startup_funding.get(currName, 0) + 1
    
ans = sorted(startup_funding.items(), key = lambda x : x[1], reverse = True)
ans = ans[:5]
```

```python
for i in range(5):
    print(ans[i][0], ans[i][1])
```

### Creating a Bar Graph for the top 5 startups

```python
startup = []
number_of_fundings = []
for i in range(5):
    startup.append(ans[i][0])
    number_of_fundings.append(ans[i][1])
    
```

```python
plt.bar(startup, number_of_fundings, color = 'teal', width = 0.5)
plt.xticks(rotation = 40)
plt.yticks(np.arange(10))
plt.xlabel('Startup Names')
plt.ylabel('Number of Fundings')
plt.title('Startups with number of fundings in each')
plt.grid()
plt.show()
```
